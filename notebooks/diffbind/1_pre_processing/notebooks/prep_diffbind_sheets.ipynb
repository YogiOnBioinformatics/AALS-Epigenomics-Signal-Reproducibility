## Glob for Files

```python
new_bam_files = glob.glob("/pool/data/globus/fraenkel_internal/AALS_signal_reproducibility/passed_qc/2_bam/*.bam")
new_peak_files = glob.glob("/pool/data/globus/fraenkel_internal/AALS_signal_reproducibility/passed_qc/4_peaks/*.narrowPeak.gz")

old_bam_files = glob.glob("/pool/data/globus/answer_als_epigenomics/released/**/*.bam", recursive=True) + glob.glob("/pool/data/globus/answer_als_epigenomics/staged/**/*.bam", recursive=True)
old_peak_files = glob.glob("/pool/data/globus/answer_als_epigenomics/released/**/*.narrowPeak.gz", recursive=True) + glob.glob("/pool/data/globus/answer_als_epigenomics/staged/**/*.narrowPeak.gz", recursive=True)
```

## Get list of Unique Participants (New & Old)

```python
# get participant ids and uniq them 
new_participant_ids = list(set([ "{}-{}".format( file.split("/")[-1].split("-")[0] , file.split("/")[-1].split("-")[1] ) for file in new_bam_files ]))
old_participant_ids = list(set([ "{}-{}".format( file.split("/")[-1].split("-")[0] , file.split("/")[-1].split("-")[1] ) for file in old_bam_files ]))

# don't compare AE8 samples
new_participant_ids.remove("CTRL-NEUEU392AE8")
old_participant_ids.remove("CTRL-NEUEU392AE8")
```

```python
new_participant_ids
old_participant_ids
```

## Find Intersection of Participants for Comparison

```python
# get intersection of IDs
participants_for_comparison = list(set(new_participant_ids).intersection(set(old_participant_ids)))
participants_for_comparison
```

## Subset for Relevant BAM and Peak Files

```python
diffbind_bam_files=[]
diffbind_peak_files=[]

for participant in participants_for_comparison: 
    new_bam_files = glob.glob("/pool/data/globus/fraenkel_internal/AALS_signal_reproducibility/passed_qc/2_bam/*{}*.bam".format(participant))
    new_peak_files = glob.glob("/pool/data/globus/fraenkel_internal/AALS_signal_reproducibility/passed_qc/4_peaks/*{}*.narrowPeak.gz".format(participant))

    old_bam_files = glob.glob("/pool/data/globus/answer_als_epigenomics/released/**/*{}*.bam".format(participant), recursive=True) + glob.glob("/pool/data/globus/answer_als_epigenomics/staged/**/*{}*.bam".format(participant), recursive=True)
    old_peak_files = glob.glob("/pool/data/globus/answer_als_epigenomics/released/**/*{}*.narrowPeak.gz".format(participant), recursive=True) + glob.glob("/pool/data/globus/answer_als_epigenomics/staged/**/*{}*.narrowPeak.gz".format(participant), recursive=True)
    
    assert len(new_bam_files)==1 and len(new_peak_files)==1 and len(old_bam_files)==1 and len(old_peak_files)==1, print(participant)
    
    diffbind_bam_files.extend(new_bam_files)
    diffbind_peak_files.extend(new_peak_files)
    diffbind_bam_files.extend(old_bam_files)
    diffbind_peak_files.extend(old_peak_files)
```

```python
len(diffbind_bam_files)
len(diffbind_peak_files)

diffbind_bam_files
diffbind_peak_files
```

# Prep Sample Sheet

```python
sample_id=[]
condition=[]
treatment=[]
bamReads=[]
peaks=[]

for i in range(0,len(diffbind_bam_files)): 
    
    tmp_bam_file = diffbind_bam_files[i]
    tmp_peak_file = diffbind_peak_files[i]
    
    if "fraenkel_internal" in tmp_bam_file: 
        tmp_treatment = "New"
    elif "answer_als_epigenomics" in tmp_bam_file: 
        tmp_treatment = "Old"
            
    parts = tmp_bam_file.split("/")[-1].split("_")[0].split("-")
    
    sample_id.append("{}-{}-{}".format(parts[0],parts[1], tmp_treatment))
    
    print("{}-{}".format(parts[0],parts[1]) + "\t" + "-".join(tmp_bam_file.split("/")[-1].split("_")[0].split("-")[0:2]) )
    
    assert "{}-{}".format(parts[0],parts[1]) == "-".join(tmp_bam_file.split("/")[-1].split("_")[0].split("-")[0:2])
    
    treatment.append(tmp_treatment)
    
    condition.append(parts[0])
    
    bamReads.append(tmp_bam_file)
    
    peaks.append(tmp_peak_file)
    
```

```python
sample_sheet = pandas.DataFrame({"SampleID":sample_id, "Condition":condition, "bamReads":bamReads, "Peaks": peaks, "PeakCaller":"narrow"})
sample_sheet
```
